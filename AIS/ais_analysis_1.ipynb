```python
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as mpl

```

```python
df = pd.read_csv('AIS_2023_01_03.csv')
df['BaseDateTime']
```

```python
df.head()
```

```python

df['BaseDateTime'] = pd.to_datetime(df['BaseDateTime'])
filtered_df = df[df['CallSign'] == 'WTC3165']
filtered_df.head(5)
```

```python

filtered_df['TimeElapsed'] = (filtered_df['BaseDateTime'] - filtered_df['BaseDateTime'].iloc[0]).dt.total_seconds()
```

```python
filtered_df.head()
```

```python
filtered_df.plot(x='TimeElapsed', y='COG', kind='line')
plt.xlabel('Time Elapsed (s)')

plt.show()
```
