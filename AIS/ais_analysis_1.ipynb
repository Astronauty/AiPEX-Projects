```python
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AIS_172779598539966953_2098-1727795986074.csv')
df['BaseDateTime']
```

```python
df.head()
```

```python

df['BaseDateTime'] = pd.to_datetime(df['BaseDateTime'])
# filtered_df = df[df['VesselType'] == 30.0]
filtered_df = df[df['CallSign'] =='WCV4687']

filtered_df.head(5)
```

```python
filtered_df['TimeElapsed'] = (filtered_df['BaseDateTime'] - filtered_df['BaseDateTime'].iloc[0]).dt.total_seconds()
```

```python
filtered_df.head()
```

```python
filtered_df['LAT_derivative'] = filtered_df['LAT'].diff()

filtered_df.plot(x='TimeElapsed', y='LAT', kind='line')
filtered_df.plot(x='TimeElapsed', y='LON', kind='line')

plt.xlabel('Time Elapsed (s)')


plt.show()
```
